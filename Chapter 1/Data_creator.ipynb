```python
import numpy as np
#import signatory
import pandas as pd
from generation_process import generate_cox_process
import birdepy as bd
from scipy.stats import poisson
import torch
import matplotlib.pyplot as plt; mpl.rcParams['text.usetex'] = True
```

```python
b = 0
B0 = poisson.rvs(mu=100, loc=0, size=1, random_state=None)[0]
d = np.log(2)/4
alpha = B0*d
jump_times = np.array([300*i for i in range(7)])
#generate_cox_process(0,d)
traj = bd.simulate.discrete([b,d,alpha], 'linear-migration', z0 = B0, times = jump_times, method = 'gwa')
plt.plot(traj)
```

```python
############### Generating then saving benign for multiple parameters
b=0
jump_times = np.array([300*i for i in range(8)])
data = np.empty([8,1000])
for d in [np.log(2)/i for i in range(4,8)]:
    for i in range(1000):
        B0 = poisson.rvs(mu=100, loc=0, size=1, random_state=None)[0]
        alpha = B0*d
        biom_traj= bd.simulate.discrete([b,d,alpha], 'linear-migration', z0 = B0, times = jump_times, method = 'gwa')
        while biom_traj[-1] == 0:
            biom_traj= bd.simulate.discrete([b,d,alpha], 'linear-migration', z0 = B0, times = jump_times, method = 'gwa')
        data[:,i] = np.array(biom_traj)
        print('b='+str(np.round(d,2))+' and i='+str(i))
    mat_biom_b = pd.DataFrame(data)
    file = './Simul/benign_dwi_b'+str(np.round(d,2))+'.csv'
    mat_biom_b.to_csv(file, sep = ';')
```

```python
############### Generating then saving malignant for multiple parameters


for b in [0.07,0.1,0.15,0.2,0.25]:
    for r in [0.001,0.002,0.004,0.007,0.01,0.015]:
        d = b-r
        if d>0:
            data = np.empty([7,1000])
            for i in range(1000):
                biom_traj, _, _ = generate_cox_process(b,d,transition=True)
                while biom_traj[-1,-1,-1] == 0:
                    biom_traj, transition_time, _ = generate_cox_process(b,d)
                data[:,i] = biom_traj[:,:,1]
                print('b='+str(b)+' and d='+str(d)+' and i='+str(i))
            mat_biom_b = pd.DataFrame(data)
            file = './Simul/mal_'+str(b)+'_r'+str(r)+'.csv'
            mat_biom_b.to_csv(file, sep = ';')
```
